```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('fish.csv')
print("First 5 rows of the dataset:")
print(data.head())

print("\nDataset Information:")
print(data.info())

print("\nDescriptive Statistics:")
print(data.describe())

print("\nMissing Values:")
print(data.isnull().sum())

plt.figure(figsize=(10, 6))
sns.countplot(x='Species', data=data, palette='Set2')
plt.title('Distribution of Fish Species')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Length3', y='Weight', hue='Species', palette='Set1', data=data)
plt.title('Weight vs Cross Length (Length3) by Species')
plt.xlabel('Cross Length (cm)')
plt.ylabel('Weight (g)')
plt.show()

plt.figure(figsize=(10, 6))
numeric_data = data.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_data.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Fish Dataset')
plt.show()

sns.pairplot(data, hue='Species', palette='Set1')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='Weight', data=data, palette='Set3')
plt.title('Boxplot of Weight by Species')
plt.xticks(rotation=45)
plt.show()
```
